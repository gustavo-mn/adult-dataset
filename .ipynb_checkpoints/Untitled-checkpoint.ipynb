```python
# Leitura dos Datasets desbalanceados

# Conjuntos de treinamento

# Conjunto 1
conjunto1_train        = pd.read_csv('Data/datasets-desbalanceados/train_data_1.csv')
conjunto1_train_target = pd.read_csv('Data/datasets-desbalanceados/train_target_1.csv')

# Conjunto 2
conjunto2_train        = pd.read_csv('Data/datasets-desbalanceados/train_data_2.csv')
conjunto2_train_target = pd.read_csv('Data/datasets-desbalanceados/train_target_2.csv')

# Conjunto 3
conjunto3_train        = pd.read_csv('Data/datasets-desbalanceados/train_data_3.csv')
conjunto3_train_target = pd.read_csv('Data/datasets-desbalanceados/train_target_3.csv')

# Conjunto 4
conjunto4_train        = pd.read_csv('Data/datasets-desbalanceados/train_data_4.csv')
conjunto4_train_target = pd.read_csv('Data/datasets-desbalanceados/train_target_4.csv')

# Conjuntos de teste
# Conjunto 1
conjunto1_test        = pd.read_csv('Data/datasets-desbalanceados/test_data_1.csv')
conjunto1_test_target = pd.read_csv('Data/datasets-desbalanceados/test_target_1.csv')

# Conjunto 2
conjunto2_test        = pd.read_csv('Data/datasets-desbalanceados/test_data_2.csv')
conjunto2_test_target = pd.read_csv('Data/datasets-desbalanceados/test_target_2.csv')

# Conjunto 3
conjunto3_test        = pd.read_csv('Data/datasets-desbalanceados/test_data_3.csv')
conjunto3_test_target = pd.read_csv('Data/datasets-desbalanceados/test_target_3.csv')

# Conjunto 4
conjunto4_test        = pd.read_csv('Data/datasets-desbalanceados/test_data_4.csv')
conjunto4_test_target = pd.read_csv('Data/datasets-desbalanceados/test_target_4.csv')
```

```python
# Lista de csvs, para facilitar qualquer loop
df_data        = [conjunto1_train, conjunto2_train ,conjunto3_train ,conjunto4_train,
                   conjunto1_test, conjunto2_test  ,conjunto3_test  ,conjunto4_test]

df_target      = [conjunto1_train_target, conjunto2_train_target, conjunto3_train_target, conjunto4_train_target,
                  conjunto1_test_target, conjunto2_test_target , conjunto3_test_target , conjunto4_test_target]

# Remoção das colunas ids
for df_index in range(0,8):
    df_data[df_index].drop('id',axis=1,inplace=True)
    df_target[df_index].drop('id',axis=1,inplace=True)

# Codificação das features categóricas, utilizando get_dummies

def get_dummies_features(df):
    ''' Função utilizada para a codificação de features categóricas de um dataframe em features binárias'''
    output = pd.DataFrame(index = df.index)

    # Investigate each feature column for the data
    for col, col_data in df.iteritems():

        if col_data.dtype == object:
            col_data = pd.get_dummies(col_data, prefix = col)  
        
        output = output.join(col_data)
    
    return output

# Convertendo os dataframes com dummies :

# Conjunto de treinamento
conjunto1_train = get_dummies_features(conjunto1_train)
conjunto2_train = get_dummies_features(conjunto2_train)
conjunto3_train = get_dummies_features(conjunto3_train)
conjunto4_train = get_dummies_features(conjunto4_train)

# Conjunto de teste
conjunto1_test = get_dummies_features(conjunto1_test)
conjunto2_test = get_dummies_features(conjunto2_test) 
conjunto3_test = get_dummies_features(conjunto3_test)
conjunto4_test = get_dummies_features(conjunto4_test)

# # Conversão das variáveis dos dataframes em dummies:
# for index, dataframe in enumerate(df_data):
#     df_data[index] = get_dummies_features(dataframe)
    
# # OBS: Neste método são criadas cópias dos dataframes! Por enquanto, o anterior é utilizado enquanto não descubro como passar como ref.
```

```python
# Módulos dos classificadores
from sklearn.neighbors import KNeighborsClassifier   as KNN
from sklearn.ensemble  import RandomForestClassifier as RF

clf_knn = KNN()
clf_RF  = RF()

# Teste base: Verificando o funcionamento correto do modelo para as configurações padrões: K = 5
clf_knn.fit(conjunto1_train,conjunto1_train_target.earnings)
prediction_conjunto1 = clf_knn.predict(conjunto1_test)
print('Acurácia para o conjunto 1: {:.2f}'.format(accuracy_score(conjunto1_test_target.earnings,prediction_conjunto1,normalize=True)))

clf_knn.fit(conjunto2_train,conjunto2_train_target.earnings)
prediction_conjunto2 = clf_knn.predict(conjunto2_test)
print('Acurácia para o conjunto 2: {:.2f}'.format(accuracy_score(conjunto2_test_target.earnings,prediction_conjunto2,normalize=True)))

clf_knn.fit(conjunto3_train,conjunto3_train_target.earnings)
prediction_conjunto3 = clf_knn.predict(conjunto3_test)
print('Acurácia para o conjunto 3: {:.2f}'.format(accuracy_score(conjunto3_test_target.earnings,prediction_conjunto3,normalize=True)))

clf_knn.fit(conjunto4_train,conjunto4_train_target.earnings)
prediction_conjunto4 = clf_knn.predict(conjunto4_test)
print('Acurácia para o conjunto 4: {:.2f}'.format(accuracy_score(conjunto4_test_target.earnings,prediction_conjunto4,normalize=True)))
```

```python
# Importando o keras
import keras
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.callbacks import TensorBoard
from time import time
from sklearn.model_selection import StratifiedKFold

train  = conjunto2_train.values
target = (conjunto2_train_target.earnings.replace(-1,0)).values

tam = conjunto2_train.shape[1]

# Exemplo de rede - teste básico

# Criação de uma rede

param_range = np.logspace(start=5,stop=5,base=2)

n_splits = 2

for n_neuronios in param_range:

    model = Sequential()
    model.add(Dense(n_neuronios, input_shape=(37,), activation='sigmoid'))
    # model.add(Dense(100, activation='sigmoid'))
    model.add(Dense(1, activation='sigmoid'))

    # Problema de classificação binária,
    model.compile(optimizer='NAdam',
                  loss='binary_crossentropy',
                  metrics=['accuracy'])


    tbCallBack = TensorBoard(log_dir='./Graph', histogram_freq=0,write_graph=True, write_images=True)
#early = keras.callbacks.EarlyStopping(monitor='val_loss', min_delta=0, patience=0, verbose=0, mode='auto', baseline=None)

    kfold = StratifiedKFold(n_splits=n_splits,shuffle=True,random_state=0)
    
#     model.fit(train[idx_treino], target[idx_treino], epochs=100, batch_size=16, callbacks=[tbCallBack])

    train_scores_acc = np.zeros(n_splits)
    val_scores_acc = np.zeros(n_splits)
    
    train_scores_f1 = np.zeros(n_splits)
    val_scores_f1 = np.zeros(n_splits)
    
    train_scores_roc = np.zeros(n_splits)
    val_scores_roc = np.zeros(n_splits)
    
    n_rodada = 0
    for idx_treino,idx_validacao in kfold.split(train,target):
        
        # Treinamento da rede
        model.fit(train[idx_treino], target[idx_treino], epochs=50, batch_size=16, callbacks=[tbCallBack])
        
        # Cálculo das medidas
        train_predictions_proba = model.predict(train[idx_treino])
        val_predictions_proba = model.predict(conjunto2_test.values[idx_validacao])
        
        train_predictions_labels = np.zeros(train_predictions_proba[:,0].shape[0])
        for i in range(0,train_predictions_proba[:,0].shape[0]):
            train_predictions_labels[i] = 1 if (train_predictions_proba[:,0][i] > 0.5) else 0
            
        val_predictions_labels = np.zeros(val_predictions_proba[:,0].shape[0])
        for i in range(0,val_predictions_proba[:,0].shape[0]):
            val_predictions_labels[i] = 1 if (val_predictions_proba[:,0][i] > 0.5) else 0
            
        train_scores_acc[n_rodada] = accuracy_score(target[idx_treino],train_predictions_labels)
        val_scores_acc[n_rodada] = accuracy_score(target[idx_validacao],val_predictions_labels)
        
        train_scores_f1[n_rodada] = f1_score(target[idx_treino],train_predictions_labels)
        val_scores_f1[n_rodada] = f1_score(target[idx_validacao],val_predictions_labels)
        
        train_scores_roc[n_rodada] = roc_score(target[idx_treino],train_predictions_proba)
        val_scores_roc[n_rodada] = roc_score(target[idx_validacao],val_predictions_proba)
        
    
```
